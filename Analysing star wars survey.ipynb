# Star Wars Survey

## Objective
To find answers to questions like :-  
* Which is the most liked star wars part/movie ?
* Which is the most viewed star wars part/movie ?
* Does the answers to above questions change for males and females ?

**Data Dictionary**  

RespondentID - An anonymized ID for the respondent (person taking the survey)  
Gender - The respondent's gender  
Age - The respondent's age  
Household Income - The respondent's income  
Education - The respondent's education level  
Location (Census Region) - The respondent's location  
Have you seen any of the 6 films in the Star Wars franchise? - Has a   Yes or No response   
Do you consider yourself to be a fan of the Star Wars film franchise? -   Has a Yes or No response  

```python
import numpy as np
```

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]
```

```python
star_wars.iloc[:, 1:3]
```

```python
yes_no = {'Yes':True, 'No': False}
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python
star_wars.iloc[:, 1:3]
```

```python
star_wars.iloc[:, 3:9]
```

```python
mapper = {
    'Star Wars: Episode I  The Phantom Menace' : True,
    np.NaN : False
    }

star_wars['Which of the following Star Wars films have you seen? Please select all that apply.'] = star_wars['Which of the following Star Wars films have you seen? Please select all that apply.'].map(mapper)
```

```python
dic_ = ['Star Wars: Episode I  The Phantom Menace' , 'Star Wars: Episode II  Attack of the Clones', 'Star Wars: Episode III  Revenge of the Sith', 'Star Wars: Episode IV  A New Hope', 'Star Wars: Episode V The Empire Strikes Back', 'Star Wars: Episode VI Return of the Jedi']
cols = [col for col in star_wars.columns[3:9]]
```

```python
for i in range(1, 6):
    mapper = {
        dic_[i] : True,
        np.NaN : False
    }
    star_wars[cols[i]] = star_wars[cols[i]].map(mapper)
```

```python
star_wars.iloc[:, 3:9]
```

### Doind analysis on sub-groups
